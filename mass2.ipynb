Absorption and stripping




```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for X
X = np.linspace(0, 0.0012, 1000)  # Dense grid with 1000 points

# Solve for Y using the given equation
Y = (87.6 * X) / (1 - 87.6 * X)

# Plot the graph
plt.figure(figsize=(8, 6))
plt.plot(X, Y, label='Y(X+1) = 87.6*X(Y+1)', color='blue')

# Add grid, title, and labels
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.minorticks_on()  # Enable minor ticks for more grid points
plt.title('Precise Grid Plot of Y(X+1) = 87.6*X(Y+1)', fontsize=14)
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)
plt.legend()
plt.xlim(0, 0.0012)
plt.ylim(0, 0.12)

# Show the plot
plt.show()
```
