# Image Enhancement for Text

This notebook demonstrates a Python approach to enhance images that contain text. The code loads an image, converts it to grayscale, applies histogram equalization and a sharpening filter to improve the text clarity.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def load_and_display(image_path):
    # Load image in BGR mode
    img = cv2.imread(image_path)
    if img is None:
        raise ValueError(f'Image not found: {image_path}')
    # Convert to RGB for display
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(); plt.imshow(img_rgb); plt.title('Original Image'); plt.axis('off'); plt.show()
    return img

def enhance_image(img):
    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Contrast Enhancement using Histogram Equalization
    hist_eq = cv2.equalizeHist(gray)

    # Sharpening using Unsharp Masking
    blurred = cv2.GaussianBlur(hist_eq, (0,0), 3)
    sharp = cv2.addWeighted(hist_eq, 1.5, blurred, -0.5, 0)

    # Display the enhanced image
    plt.figure(); plt.imshow(sharp, cmap='gray'); plt.title('Enhanced Image'); plt.axis('off'); plt.show()
    return sharp

image_path = 'E:\ONEDRIVE\OneDrive - BUET\Study\Heat transfer\EZRA KLEIN cODE\image_2025-04-24_12-31-09.png'  # update with your image path
img = load_and_display(image_path)
enhanced = enhance_image(img)
```
