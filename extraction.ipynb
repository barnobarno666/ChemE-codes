```python
import ternary
import matplotlib.pyplot as plt

# Define the scale (e.g., 100 for percentage-based data)
scale = 100

# Equilibrium data for the acetone-water-chloroform system
# Format: (Chloroform, Water, Acetone)
chloroform_phase = [
    (0.81, 99.19, 0.00),  # Point 1
    (1.23, 82.97, 15.80),  # Point 2
    (1.29, 73.11, 25.60),  # Point 3
    (1.71, 62.29, 36.00),  # Point 4
    (5.10, 45.60, 49.30),  # Point 5
    (9.80, 34.50, 55.70),  # Point 6
]

water_phase = [
    (99.50, 0.50, 0.00),  # Point 1
    (70.00, 1.30, 28.70),  # Point 2
    (55.70, 2.20, 42.10),  # Point 3
    (42.90, 4.40, 52.70),  # Point 4
    (28.40, 10.30, 61.30),  # Point 5
    (20.40, 18.60, 61.00),  # Point 6
]

# Create a ternary plot
figure, tax = ternary.figure(scale=scale)

# Draw the boundary and gridlines
tax.boundary(linewidth=1.5)
tax.gridlines(multiple=10, color="black", linestyle="--", linewidth=0.5)

# Plot the equilibrium data
tax.scatter(chloroform_phase, marker='o', color='blue', label="Chloroform Phase")
tax.scatter(water_phase, marker='x', color='red', label="Water Phase")

# Set axis labels and title
tax.set_title("Acetone-Water-Chloroform Ternary Diagram", fontsize=14)
tax.left_axis_label("Mass fraction Acetone (B)", fontsize=12, offset=0.14)
tax.right_axis_label("Mass fraction Water (A)", fontsize=12, offset=0.14)
tax.bottom_axis_label("Mass fraction Chloroform (C)", fontsize=12, offset=0.14)

# Add legend
tax.legend()

# Set ticks
tax.ticks(axis='lbr', linewidth=1, multiple=10, offset=0.02)

# Background color
tax.set_background_color(color="whitesmoke", alpha=0.7)

# Remove default Matplotlib Axes
tax.clear_matplotlib_ticks()
tax.get_axes().axis('off')

# Show the plot
plt.show()
```
